```python
data = pd.read_csv('lenta-ru-news.csv.bz2')
data.head()
```

```python
data = data.drop(['url', 'title', 'topic', 'date'], axis=1)
data.columns = ['text', 'topic']
```

```python
data.isna().sum()
```

```python
data = data[data.notna().all(axis=1)]
```

```python
len(data)
```

```python
original2new = {}

with open('topics.tsv') as file:
    for line in file:
        new_topic, original_topics = line.removesuffix('\n').split('\t')
        for original_topic in original_topics.split(','):
            original2new[original_topic] = new_topic

original2new
```

```python
data_mail = pd.read_excel('data.mail.xlsx')
```

```python
data = data[data.topic.isin(list(original2new))]
len(data)
```

```python
data.topic = data.topic.replace(original2new)
```

```python
data.head()
```

```python
data.topic.unique()
```

```python
len(data.text.unique())
```

```python
data_mail = pd.read_excel('data.mail.xlsx', sheet_name=['health.mail', 'deti.mail'])
data_mail['health.mail'].head()
```

```python
data_mail = pd.concat([data_mail['health.mail'], data_mail['deti.mail']])
```

```python
data_mail.head()
```

```python
data_mail = data_mail.drop(['URL', 'Title', 'Description'], axis=1)
data_mail.columns = ['text', 'topic']
```

```python
data_mail.topic.unique()
```

```python
data_mail = data_mail[data_mail.topic.isin(['Семья', 'Питание', 'Образ жизни'])]
```

```python
data_mail.topic.unique()
```

```python
merged_data = pd.concat([data, data_mail])
len(merged_data)
```

```python
merged_data.topic = merged_data.topic.replace({'Питание': 'Здоровье',
                                               'Образ жизни': 'Здоровье'})
```

```python
counts = merged_data.topic.value_counts()
```

```python
indices = counts[counts > 2000].index
```

```python
merged_data = pd.concat([merged_data[merged_data.topic.isin(indices)].groupby('topic').sample(2000),
                         merged_data[~merged_data.topic.isin(indices)]])
```

```python
merged_data.topic.value_counts()
```

```python
import re

import nltk
from nltk.tokenize import WordPunctTokenizer
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
```

```python
ru_stopwords = stopwords.words('russian')
tokenize = WordPunctTokenizer().tokenize
```

```python
def preprocess_text(text: str) -> str:
    pattern = "[A-Za-z0-9#$%&'()*+/:;<=>@[\]^_`{|}~—\"\-]+"
    text = re.sub(pattern, ' ', text)

    return ' '.join([token for token in tokenize(text) if token not in ru_stopwords])
```

```python
merged_data.text = merged_data.text.map(preprocess_text)
```

```python
from sklearn.model_selection import train_test_split
```

```python
merged_data = merged_data.drop_duplicates()
len(merged_data)
```

```python
train_data, test_data = train_test_split(merged_data, shuffle=True, stratify=merged_data.topic)
train_data, val_data = train_test_split(train_data, shuffle=True, stratify=train_data.topic)
```

```python
len(train_data), len(test_data), len(val_data)
```

```python
train_data.to_csv('train.csv.gz', compression='gzip', index=False)
test_data.to_csv('test.csv.gz', compression='gzip', index=False)
val_data.to_csv('val.csv.gz', compression='gzip', index=False)
```
